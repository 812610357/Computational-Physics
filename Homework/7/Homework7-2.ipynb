```python
import numpy as np
from pandas import DataFrame as DF
import matplotlib.pyplot as plt
```

# 编程 6.3-1(a)

```python
def Euler_method(f, a, b, h, y_0):
    n = int((b - a)/h)
    d = len(y_0)
    
    t = np.linspace(a, b, n+1)
    w = np.zeros((d, n+1))
    w[:, 0] = y_0
    
    for i in range(n):
        w[:, i+1] = w[:, i] + h*(f(t[i], w[:, i]))
        
    return t, w
```

```python
a, b = 0, 1
h = [0.1,0.01]
y_0 = np.array([1,0])

def f(t, y):
    y1,y2=y
    return np.array([y1+y2,-y1+y2])

def s(t):
    return np.array([np.exp(t)*np.cos(t),-np.exp(t)*np.sin(t)])

t1, w1 = Euler_method(f, a, b, h[0], y_0)
t2, w2 = Euler_method(f, a, b, h[1], y_0)
real=s(np.arange(a,b+h[1],h[1]))

error1=np.abs(w1[:,-1]-s(1))
error2=np.abs(w2[:,-1]-s(1))

DF({"y": ['y1','y2'], "error(h=0.1)": error1, "error(h=0.01)": error2}).set_index('y')[['error(h=0.1)', 'error(h=0.01)']]
```

```python
plt.plot(t1,w1[0],label='h=0.1')
plt.plot(t2,w2[0],label='h=0.01')
plt.plot(t2,real[0],'--',label='real')
plt.legend()
plt.title('y1')
plt.xlabel('t')
plt.ylabel('y')
```

```python
plt.plot(t1,w1[1],label='h=0.1')
plt.plot(t2,w2[1],label='h=0.01')
plt.plot(t2,real[1],'--',label='real')
plt.legend()
plt.title('y2')
plt.xlabel('t')
plt.ylabel('y')
```

# 编程 6.3-2(a)

```python
def Trapezoid_method(f, a, b, h, y_0):
    n = int((b - a)/h)
    d = len(y_0)
    
    t = np.linspace(a, b, n+1)
    w = np.zeros((d, n+1))
    w[:, 0] = y_0
    
    for i in range(n):
        w[:, i+1] = w[:, i] + (f(t[i], w[:, i]) + f(t[i] + h, w[:, i] + h*f(t[i], w[:, i])))*(h/2)
        
    return t, w
```

```python
a, b = 0, 1
h = [0.1,0.01]
y_0 = np.array([1,0])

def f(t, y):
    y1,y2=y
    return np.array([y1+y2,-y1+y2])

def s(t):
    return np.array([np.exp(t)*np.cos(t),-np.exp(t)*np.sin(t)])

t1, w1 = Trapezoid_method(f, a, b, h[0], y_0)
t2, w2 = Trapezoid_method(f, a, b, h[1], y_0)
real=s(np.arange(a,b+h[1],h[1]))

error1=np.abs(w1[:,-1]-s(1))
error2=np.abs(w2[:,-1]-s(1))

DF({"y": ['y1','y2'], "error(h=0.1)": error1, "error(h=0.01)": error2}).set_index('y')[['error(h=0.1)', 'error(h=0.01)']]
```

```python
plt.plot(t1,w1[0],label='h=0.1')
plt.plot(t2,w2[0],label='h=0.01')
plt.plot(t2,real[0],'--',label='real')
plt.legend()
plt.title('y1')
plt.xlabel('t')
plt.ylabel('y')
```
