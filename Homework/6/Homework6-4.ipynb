```python
import numpy as np
from pandas import DataFrame as DF
import matplotlib.pyplot as plt
```

# 编程 6.1-2(b)

```python
def Euler_method(f, a, b, h, y_0):
    n = int((b - a)/h)
    t = np.linspace(a, b, n+1)
    w = np.zeros(n+1)
    w[0] = y_0
    
    for i in range(n):
        w[i+1] = w[i] + h*(f(t[i], w[i]))
        
    return t, w
```

```python
a, b = 0, 1
y_0 = 1
h = [0.1,0.05,0.025]

t=list([])
w=list([])
true_value=list([])
error=list([])

def f(t, y):
    return t**2*y

def exact_sol(t):
    return np.exp(1/3*t**3)

for i in range(len(h)):
    tt, ww = Euler_method(f, a, b, h[i], y_0)
    t.append(tt)
    w.append(ww)
    true_value=exact_sol(tt)
    error.append(abs(true_value - ww))

for i in range(len(h)):
    plt.plot(t[i],w[i],label='h = '+str(h[i]))

plt.plot(t[-1],true_value,label='true value')
plt.legend()
plt.xlabel('t')
plt.ylabel('y')
```

# 编程 6.1-4(b)

```python
t=list([])
w=list([])
true_value=list([])
error=list([])

def h(k):
    return 0.1*2**(-k)

for k in range(6):
    tt, ww = Euler_method(f, a, b, h(k), y_0)
    t.append(tt)
    w.append(ww)
    true_value=exact_sol(tt)
    error.append(abs(true_value - ww))

for k in range(6):
    plt.scatter(h(k),error[k][-1])

plt.xscale("log")
plt.yscale("log")
plt.xlabel('h')
plt.ylabel('error')
```

# 编程 6.2-2(b)

```python
def Explicit_trapezoid_method(f, a, b, h, y_0):
    n = int((b - a)/h)
    t = np.linspace(a, b, n+1)
    w = np.zeros(n+1)
    w[0] = y_0
    
    for i in range(n):
        w[i+1] = w[i] + (f(t[i], w[i]) + f(t[i] + h, w[i] + h*f(t[i], w[i])))*(h/2)
        
    return t, w
```

```python
h = [0.1,0.05,0.025]

t=list([])
w=list([])
true_value=list([])
error=list([])

def f(t, y):
    return t**2*y

def exact_sol(t):
    return np.exp(1/3*t**3)

for i in range(len(h)):
    tt, ww = Explicit_trapezoid_method(f, a, b, h[i], y_0)
    t.append(tt)
    w.append(ww)
    true_value=exact_sol(tt)
    error.append(abs(true_value - ww))

for i in range(len(h)):
    plt.plot(t[i],w[i],label='h = '+str(h[i]))

plt.plot(t[-1],true_value,label='true value')
plt.legend()
plt.xlabel('t')
plt.ylabel('y')
```
