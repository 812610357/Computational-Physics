```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sin(x)-np.cos(x)

def ff(x):
    return np.cos(x)+np.sin(x)

def df(f, x, h):
    return (f(x+h) - f(x-h))/(2*h)

print('h     f\'(0)      error')
plt.xscale("log")
plt.yscale("log")
for i in range(1,13):
    h=10**(-i)
    print('%.0e %.8f %e'%(h, df(f,0,h), np.abs(df(f,0,h)-ff(0))))
    plt.scatter(h,np.abs(df(f,0,h)-ff(0)),color='g')
```
