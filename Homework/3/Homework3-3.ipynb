# 2.5-2

```python
def Jacobi(A,b,x,r):
    global steps
    D=np.diag(np.diag(A))
    Di=np.linalg.inv(D)
    LU=A-D
    error=1
    while error>1e-3:
        x=np.dot(Di,b-np.dot(LU,x))
        error=np.linalg.norm(x-r,ord=np.inf)
        steps+=1
    return x
```

```python
def newA(n):
    A=np.diag(2*np.ones(n))
    B=np.diag(np.ones(n-1))
    A=A+np.pad(B,((1,0),(0,1)))+np.pad(B,((0,1),(1,0)))
    return A

def newb(n):
    b=np.zeros((n,1))
    b[0,0]=1
    b[-1,0]=-1
    return b

steps=0
n=100
A=newA(n)
b=newb(n)
x0=np.zeros((n,1))
r=np.mod(np.array([np.linspace(1,n,n)]).T,2)*2-1
x=Jacobi(A,b,x0,r)
print(x)
print('steps',steps)
print('BE',np.linalg.norm(b-A.dot(x),ord=np.inf))
```

# 2.5-6(a)

```python
def Gauss_Seidel(A,b,x_i,r):
    global steps
    D=np.diag(np.diag(A))
    Di=np.linalg.inv(D)
    L = np.tril(A) - D
    U = np.triu(A) - D
    error=1
    x_ii=x_i
    while error>1e-3:
        for k in range(b.size):
            x_ii[k] =np.dot(Di[k,k], (b[k] - np.dot(U[k],x_i) - np.dot(L[k],x_ii)))
        error=np.linalg.norm(x_ii-r,ord=np.inf)
        x_i=x_ii
        steps+=1
    return x_ii
```

```python
def newA(n):
    A=np.diag(2*np.ones(n))
    B=np.diag(np.ones(n-1))
    A=A+np.pad(B,((1,0),(0,1)))+np.pad(B,((0,1),(1,0)))
    return A

def newb(n):
    b=np.zeros((n,1))
    b[0,0]=1
    b[-1,0]=-1
    return b

steps=0
n=100
A=newA(n)
b=newb(n)
x0=np.zeros((n,1))
r=np.mod(np.array([np.linspace(1,n,n)]).T,2)*2-1
x=Gauss_Seidel(A,b,x0,r)
print(x)
print('steps',steps)
print('BE',np.linalg.norm(b-A.dot(x),ord=np.inf))
```

# 2.5-6(b)

```python
def SOR(A,b,x,r,w):
    global steps
    D=np.diag(np.diag(A))
    L = np.tril(A) - D
    U = np.triu(A) - D
    Di=np.linalg.inv(w*L+D)
    error=1
    while error>1e-3:
        x=np.dot(Di,(1-w)*np.dot(D,x)-w*np.dot(U,x))+w*np.dot(Di,b)
        error=np.linalg.norm(x-r,ord=np.inf)
        steps+=1
    return x
```

```python
def newA(n):
    A=np.diag(2*np.ones(n))
    B=np.diag(np.ones(n-1))
    A=A+np.pad(B,((1,0),(0,1)))+np.pad(B,((0,1),(1,0)))
    return A

def newb(n):
    b=np.zeros((n,1))
    b[0,0]=1
    b[-1,0]=-1
    return b

steps=0
n=100
A=newA(n)
b=newb(n)
x0=np.zeros((n,1))
r=np.mod(np.array([np.linspace(1,n,n)]).T,2)*2-1
x=SOR(A,b,x0,r,1.5)
print(x)
print('steps',steps)
print('BE',np.linalg.norm(b-A.dot(x),ord=np.inf))
```
